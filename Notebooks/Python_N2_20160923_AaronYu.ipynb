## Control Flow

- for loop

```python
countries = ['US', 'CN', 'UK','JP']
for country in countries:
    print(country)
```

- if, elif, else

```python
score = 75
if score > 90:
    print('A')
elif score > 80:
    print('B')
elif score > 60:
    print('C')
else:
    print('D')
```

- for and if combined

```python
num = [2,3,7,8,1]
for i in num:
    if i % 2 == 0:
        print('Even')
    else:
        print('Odd')
```

```python
for i in range(len(countries)):
    print('The %dth item is: %s' % (i, countries[i]))
```

    2.np.arange()

```python
## np.arange() returns evenly spaced values within a given interval. 
## For integer arguments the function is equivalent to the range function
## But returns an ndarray rather than a list
```

```python
arr4 = np.arange(10)
arr4
```

```python
arr5 = np.arange(1,3,0.1)
arr5
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,5,0.2)
y = [ele**2 for ele in x]
plt.plot(x,y)
```

```python
x_2 = range(5)
y_2 = [ele**2 for ele in x_2]
plt.plot(x_2, y_2)
```
